```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/marionomics/econometria/main/data/Student_Behaviour.csv")
df.columns = ["courses", "gender", "dept", "height", "weight", "ten_m", "twelve_m", "collegeM", "hobbies", "st", "time", "exp", "lk_deg", "prob", "sm_vg", "trav_time", "stress", "fin", "ptj"]
```

```python
df.head()
```

```python
df2 = df[["courses","gender", "height", "ten_m", "twelve_m", "collegeM", "st", "fin"]]
df2 = pd.get_dummies(df2)
df2
```

```python
df2.courses_Yes.mean()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
```

```python
model = sm.OLS.from_formula('collegeM ~ courses_Yes + gender_Female + ten_m', data = df2)
model.fit().summary()
```

```python
type(model)
```

```python
df3 = df2.copy()
```

```python
df3['predict_college'] = model.fit().predict()
df3['paso'] = np.where(df3['collegeM'] > 60.0, 1,0)
df3['predict_paso'] = np.where(df3['predict_college'] > 60.0, 1,0)
df3[['collegeM', 'predict_college', 'paso', 'predict_paso']]
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(df3['paso'], df3['predict_paso'])
```
